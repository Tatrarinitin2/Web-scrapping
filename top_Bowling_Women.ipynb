```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/bowling')
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
name=soup.find('div', class_="rankings-block__banner--name-large")
name.text
```

```python
team=soup.find('div', class_="rankings-block__banner--nationality")
team.text
```

```python
rating=soup.find('div', class_="rankings-block__banner--rating")
rating.text
```

```python
name2=soup.find('td', class_="table-body__cell rankings-table__name name")
name2.text
```

```python
team2=soup.find('span', class_="table-body__logo-text")
team2.text
```

```python
rating2=soup.find('td', class_="table-body__cell rating")
rating2.text
```

```python
NAME=[]
TEAM=[]
RATING=[]
```

```python
NAME.append(name.text)
TEAM.append(team.text.replace('\n',''))
RATING.append(rating.text)
```

```python
name3=soup.find_all('td', class_="table-body__cell rankings-table__name name")
name3[0:4]
```

```python
team3=soup.find_all('span', class_="table-body__logo-text")
team3[0:4]
```

```python
rating3=soup.find_all('td', class_="table-body__cell rating")
rating3[0:4]
```

```python
for i in name3:
    NAME.append(i.text.replace('\n',''))
```

```python
for i in team3:
    TEAM.append(i.text)
```

```python
for i in rating3:
    RATING.append(i.text)
```

```python
import pandas as pd
BW_ODI=pd.DataFrame([])
```

```python
BW_ODI['PLAYER']=NAME[0:10]
BW_ODI['TEAM']=TEAM[0:10]
BW_ODI['RATING']=RATING[0:10]
BW_ODI
```

```python
BW_ODI.to_csv
```
