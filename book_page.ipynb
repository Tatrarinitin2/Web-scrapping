```python
Book_Name=[]
Author_Name=[]
Genre=[]
Review=[]
```

```python
Book_Name.append(name.text.replace('\n',''))
Author_Name.append(aname.text.replace('\n',''))
Genre.append(genre.text.replace('\n',''))
Review.append(review.text.replace('\n',''))
```

```python
link=['https://bookpage.com/reviews/26014-gayle-tzemach-lemmon-daughters-kobani-nonfiction#.YFcaktLitPY','https://bookpage.com/reviews/26102-arik-kershenbaum-zoologists-guide-to-galaxy-nonfiction#.YFcbbdLitPY',
     'https://bookpage.com/reviews/26090-tommy-greenwald-rivals-childrens#.YFcdSNLitPY','https://bookpage.com/reviews/26072-jamie-figueroa-brother-sister-mother-explorer-fiction#.YFcbt9LitPY ']
for i in link:
    page2=requests.get(i)
    soup2=BeautifulSoup(page2.content,'html.parser')
    name=soup2.find('h1' ,class_="italic")
    aname=soup2.find('h4' ,class_="sans")
    genre=soup2.find('p', class_="genre-links")
    review=soup2.find('div', class_="span7")
    Book_Name.append(name.text.replace('\n',''))
    Author_Name.append(aname.text.replace('\n',''))
    Genre.append(genre.text.replace('\n',''))
    Review.append(review.text.replace('\n',''))
```

```python
import pandas as pd
Book_Page=pd.DataFrame([])
Book_Page['Book_Name']=Book_Name[:]
Book_Page['Author_Name']=Author_Name[:]
Book_Page['Genre']=Genre[:]
Book_Page['Review']=Review[:]
```
